# 折线图的绘制与显示

```python
# 举例:展现上海一周的天气,比如从星期一到星期日的天气温度如下

import matplotlib.pyplot as plt

# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 2. 绘制图像
plt.plot([1, 2, 3, 4, 5, 6 ,7], [17,17,18,15,11,11,13])

# 3. 保存图像，注意，不能保出现在 `plt.show()` 后面
plt.savefig('./pic/折线图.png')

# 4. 显示图像，同时释放画布的所有资源
plt.show()

```
